```python
import multiprocessing as mp
from joblib import Parallel, delayed
import time
import numpy as np
import datetime
import json
import os

# Global variables to store optimization state
# These need to be global for pickling to work on macOS
EVALUATIONS_CACHE = {}
LOG_FILE = None
TOTAL_START_TIME = None
EVAL_COUNTER = 0
N_CALLS = 0
RUN_DIR = None
VERBOSE = 1

def simulate_single_route_optimized(i, weights_dict, calculation_type='both'):
    """
    Optimized route simulation that can calculate just distance, just objective, or both
    
    Args:
        i: Index of the route in test_set
        weights_dict: Dictionary with routing weights
        calculation_type: 'distance', 'objective', or 'both'
    """
    points = test_set[i]
    result = {'route_index': i, 'success': False}
    
    try:
        # Only calculate distance-based route if needed
        if calculation_type in ['distance', 'both']:
            start_time = time.time()
            route_output_dist = sr.find_route(map_data, road_df, graph, points[0], points[1], 
                                             weights_dict, plot=False, weights_type='distance')
            dist_consumption, _, _, _, dist_current_list, _, _ = sr.return_route_data_complex(
                route_output_dist, vehicle_data, static_data, motor_eff, battery_data)
            
            result['distance_consumption'] = dist_consumption
            result['distance_currents'] = dist_current_list
            result['distance_time'] = time.time() - start_time
        
        # Only calculate objective-based route if needed
        if calculation_type in ['objective', 'both']:
            start_time = time.time()
            route_output_obj = sr.find_route(map_data, road_df, graph, points[0], points[1], 
                                            weights_dict, plot=False, weights_type='objective')
            obj_consumption, _, _, _, obj_current_list, _, _ = sr.return_route_data_complex(
                route_output_obj, vehicle_data, static_data, motor_eff, battery_data)
            
            result['objective_consumption'] = obj_consumption
            result['objective_currents'] = obj_current_list
            result['objective_time'] = time.time() - start_time
        
        result['success'] = True
        return result
    
    except Exception as e:
        print(f'Route {i} failed: {str(e)}')
        return result

def run_parallel_optimized(weights_dict=None, num_routes=None, calculation_type='both', n_jobs=None, verbose=1):
    """
    Optimized parallel execution with more flexibility
    
    Args:
        weights_dict: Dictionary with routing weights
        num_routes: Number of routes to process (None = all routes)
        calculation_type: 'distance', 'objective', or 'both'
        n_jobs: Number of parallel jobs (defaults to CPU count - 1)
        verbose: Verbosity level (0=quiet, 1=normal, 2=detailed)
    """
    start_time = time.time()
    
    # Use default weights if none provided
    if weights_dict is None:
        weights_dict = {
            "incline_weight": 4.0,
            "max_incline_weight": 4.0,
            "distance_weight": 10.0,
            "zero_start_weight": 10.0
        }
    
    # Set number of parallel jobs
    if n_jobs is None:
        n_jobs = max(1, mp.cpu_count() - 1)
    
    # Select route indices 
    if num_routes is None:
        num_routes = len(test_set)
    else:
        num_routes = min(num_routes, len(test_set))
    
    route_indices = list(range(num_routes))
    
    if verbose >= 1:
        print(f"Running {num_routes} simulations using {n_jobs} parallel jobs")
        print(f"Calculation type: {calculation_type}")
    
    # Run in parallel using joblib with configurable verbosity
    joblib_verbose = 10 if verbose >= 2 else 0
    results = Parallel(n_jobs=n_jobs, verbose=joblib_verbose)(
        delayed(simulate_single_route_optimized)(i, weights_dict, calculation_type) 
        for i in route_indices
    )
    
    # Process results
    successful_results = [r for r in results if r['success']]
    failed_routes = [r['route_index'] for r in results if not r['success']]
    
    # Calculate timing information
    total_time = time.time() - start_time
    
    if verbose >= 1:
        print(f"\nResults summary:")
        print(f"- Total execution time: {total_time:.2f} seconds")
        print(f"- Average time per route: {total_time/num_routes:.2f} seconds")
        print(f"- Successfully simulated {len(successful_results)} routes")
        print(f"- Failed to simulate {len(failed_routes)} routes")
        
        if failed_routes:
            if len(failed_routes) <= 10 or verbose >= 2:
                print(f"- Failed route indices: {failed_routes}")
            else:
                print(f"- Failed {len(failed_routes)} routes (too many to display)")
    
    # Calculate statistics based on calculation type
    if successful_results and verbose >= 1:
        if calculation_type in ['distance', 'both']:
            avg_dist_consumption = sum(r['distance_consumption'] for r in successful_results) / len(successful_results)
            if 'distance_time' in successful_results[0]:
                avg_dist_time = sum(r.get('distance_time', 0) for r in successful_results) / len(successful_results)
                print(f"Average distance calculation time: {avg_dist_time:.2f} seconds")
            print(f"Average distance-based consumption: {avg_dist_consumption:.2f}")
        
        if calculation_type in ['objective', 'both']:
            avg_obj_consumption = sum(r['objective_consumption'] for r in successful_results) / len(successful_results)
            if 'objective_time' in successful_results[0]:
                avg_obj_time = sum(r.get('objective_time', 0) for r in successful_results) / len(successful_results)
                print(f"Average objective calculation time: {avg_obj_time:.2f} seconds")
            print(f"Average objective-based consumption: {avg_obj_consumption:.2f}")
        
        if calculation_type == 'both':
            better_count = sum(1 for r in successful_results if r['objective_consumption'] < r['distance_consumption'])
            same_count = sum(1 for r in successful_results if r['objective_consumption'] == r['distance_consumption'])
            worse_count = sum(1 for r in successful_results if r['objective_consumption'] > r['distance_consumption'])
            
            print(f"\nObjective routing performed better than distance in {better_count} routes")
            print(f"Objective routing gave same result as distance in {same_count} routes")
            print(f"Objective routing performed worse than distance in {worse_count} routes")
    
    return successful_results

def format_time(seconds):
    """Format seconds into a human-readable time string"""
    if seconds < 60:
        return f"{seconds:.1f} seconds"
    elif seconds < 3600:
        minutes = seconds / 60
        return f"{minutes:.1f} minutes"
    else:
        hours = seconds / 3600
        return f"{hours:.1f} hours"

def save_cache():
    """Save the evaluations cache to disk"""
    global EVALUATIONS_CACHE, RUN_DIR
    if RUN_DIR:
        cache_file = os.path.join(RUN_DIR, "evaluations_cache.json")
        serializable_cache = {str(k): v for k, v in EVALUATIONS_CACHE.items()}
        with open(cache_file, "w") as f:
            json.dump(serializable_cache, f)

# This function must be at module level for pickling to work on macOS
def objective_function_optimized(weights_array):
    """Objective function for optimization - must be at module level"""
    global EVALUATIONS_CACHE, LOG_FILE, TOTAL_START_TIME, EVAL_COUNTER, N_CALLS, RUN_DIR, VERBOSE
    
    EVAL_COUNTER += 1
    eval_start_time = time.time()
    
    # Convert array of weights to dictionary
    weights_dict = {
        "incline_weight": weights_array[0],
        "max_incline_weight": weights_array[1],
        "distance_weight": weights_array[2],
        "zero_start_weight": weights_array[3]
    }
    
    # Create a cache key (convert weights to tuple for hashability)
    cache_key = tuple(weights_array)
    
    # Check if we have already evaluated these weights
    if cache_key in EVALUATIONS_CACHE:
        if VERBOSE >= 1:
            print(f"Using cached result for weights: {weights_dict}")
        
        # Log the cache hit
        if LOG_FILE:
            with open(LOG_FILE, "a") as f:
                f.write(f"Evaluation {EVAL_COUNTER}/{N_CALLS} - CACHE HIT - {weights_dict} - Value: {EVALUATIONS_CACHE[cache_key]}\n")
        
        return EVALUATIONS_CACHE[cache_key]
    
    # Estimate remaining time
    elapsed_time = time.time() - TOTAL_START_TIME
    if EVAL_COUNTER > 1:
        time_per_eval = elapsed_time / (EVAL_COUNTER - 1)  # Exclude current evaluation
        remaining_evals = N_CALLS - EVAL_COUNTER + 1  # Include current evaluation
        estimated_remaining = time_per_eval * remaining_evals
        
        if VERBOSE >= 1:
            print(f"\nEvaluation {EVAL_COUNTER}/{N_CALLS} - Elapsed: {format_time(elapsed_time)}, Estimated remaining: {format_time(estimated_remaining)}")
            print(f"Evaluating weights: {weights_dict}")
    else:
        if VERBOSE >= 1:
            print(f"\nEvaluation {EVAL_COUNTER}/{N_CALLS} - First evaluation")
            print(f"Evaluating weights: {weights_dict}")
    
    # Run parallel evaluation
    results = run_parallel_optimized(
        weights_dict=weights_dict,
        num_routes=None,  # Use all routes by default
        calculation_type='objective',  # Only calculate objective routing to save time
        n_jobs=max(1, mp.cpu_count() - 1),  # Use available cores
        verbose=max(0, VERBOSE-1)  # Reduce verbosity for nested calls
    )
    
    # Calculate objective value
    if not results:
        # If all routes failed, return a high penalty value
        objective_value = 1e6  # High value for minimization
    else:
        obj_consumption_sum = sum(r['objective_consumption'] for r in results)
        obj_current_sum = sum(sum(r['objective_currents']) for r in results)
        objective_value = 0.7 * obj_consumption_sum + 0.3 * obj_current_sum
    
    # Cache the result
    EVALUATIONS_CACHE[cache_key] = objective_value
    
    # Save the cache every 5 evaluations
    if EVAL_COUNTER % 5 == 0:
        save_cache()
    
    eval_time = time.time() - eval_start_time
    
    # Log the evaluation
    if LOG_FILE:
        with open(LOG_FILE, "a") as f:
            f.write(f"Evaluation {EVAL_COUNTER}/{N_CALLS} - {weights_dict} - Value: {objective_value} - Time: {eval_time:.2f}s\n")
    
    if VERBOSE >= 1:
        print(f"Evaluation {EVAL_COUNTER} complete - Value: {objective_value} - Time: {eval_time:.2f}s")
    
    return objective_value

def optimize_bayesian_parallel(
    n_jobs=None, 
    n_calls=30, 
    n_initial_points=10, 
    num_routes=None,
    verbose=1,
    log_dir="optimization_logs",
    start_weights=None  # New parameter for starting weights
):
    """
    Full-scale Bayesian optimization using parallel route evaluation
    
    Parameters:
        n_jobs: Number of parallel jobs
        n_calls: Total number of function evaluations
        n_initial_points: Number of initial points for the optimizer
        num_routes: Number of routes to use (None = all routes)
        verbose: Verbosity level (0=quiet, 1=normal, 2=detailed)
        log_dir: Directory to save logs and checkpoints
        start_weights: Dictionary with pre-optimized weights to use as starting point
    """
    from skopt import gp_minimize, space
    from skopt.callbacks import CheckpointSaver, VerboseCallback
    import os
    
    # Set global variables for the optimization
    global EVALUATIONS_CACHE, LOG_FILE, TOTAL_START_TIME, EVAL_COUNTER, N_CALLS, RUN_DIR, VERBOSE
    
    # Reset globals
    EVALUATIONS_CACHE = {}
    EVAL_COUNTER = 0
    N_CALLS = n_calls
    VERBOSE = verbose
    TOTAL_START_TIME = time.time()
    
    # Create log directory if it doesn't exist
    if not os.path.exists(log_dir):
        os.makedirs(log_dir)
    
    # Generate a timestamp for this optimization run
    timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
    run_id = f"opt_run_{timestamp}"
    RUN_DIR = os.path.join(log_dir, run_id)
    os.makedirs(RUN_DIR)
    
    # Create a log file for this run
    LOG_FILE = os.path.join(RUN_DIR, "optimization_log.txt")
    with open(LOG_FILE, "w") as f:
        f.write(f"Optimization run started at {datetime.datetime.now()}\n")
        f.write(f"Parameters: n_calls={n_calls}, n_initial_points={n_initial_points}, n_jobs={n_jobs}, num_routes={num_routes}\n")
        if start_weights:
            f.write(f"Starting with pre-optimized weights: {start_weights}\n")
        f.write("\n")
    
    # Make sure n_calls is at least as large as n_initial_points
    if n_calls < n_initial_points:
        n_calls = n_initial_points
        N_CALLS = n_calls
        print(f"WARNING: Increasing n_calls to {n_initial_points} to meet scikit-optimize requirements")
    
    # Define the search space (adjust based on your pre-optimized weights)
    if start_weights:
        # Expand the search space if the pre-optimized values are near the bounds
        incline_min = min(0.1, start_weights['incline_weight'] * 0.5)
        incline_max = max(5.0, start_weights['incline_weight'] * 1.5)
        
        max_incline_min = min(0.1, start_weights['max_incline_weight'] * 0.5)
        max_incline_max = max(8.0, start_weights['max_incline_weight'] * 1.5)
        
        distance_min = min(3.0, start_weights['distance_weight'] * 0.5)
        distance_max = max(8.0, start_weights['distance_weight'] * 1.5)
        
        zero_start_min = min(10.0, start_weights['zero_start_weight'] * 0.5)
        zero_start_max = max(25.0, start_weights['zero_start_weight'] * 1.5)
        
        refined_space = [
            (incline_min, incline_max),           # incline_weight
            (max_incline_min, max_incline_max),   # max_incline_weight
            (distance_min, distance_max),         # distance_weight
            (zero_start_min, zero_start_max)      # zero_start_weight
        ]
    else:
        # Default search space
        refined_space = [
            (0.1, 5.0),       # incline_weight
            (0.1, 8.0),       # max_incline_weight
            (3.0, 8.0),       # distance_weight
            (10.0, 25.0)      # zero_start_weight
        ]
    
    # Create starting point from pre-optimized weights if provided
    x0 = None
    if start_weights:
        x0 = [
            start_weights['incline_weight'],
            start_weights['max_incline_weight'],
            start_weights['distance_weight'],
            start_weights['zero_start_weight']
        ]
        if verbose >= 1:
            print(f"Using provided weights as starting point: {start_weights}")
            print(f"Search space adjusted to: {refined_space}")
    
    # Implement the callback for checkpointing
    checkpoint_path = os.path.join(RUN_DIR, "checkpoint.pkl")
    checkpoint_saver = CheckpointSaver(checkpoint_path, compress=9)
    
    # Run the optimization
    if verbose >= 1:
        print(f"Starting optimization with {n_calls} evaluations on {num_routes if num_routes else 'all'} routes")
        print(f"Logs and checkpoints will be saved in: {RUN_DIR}")
    
    callback = VerboseCallback(n_total=n_calls) if verbose >= 1 else None
    callbacks = [checkpoint_saver]
    if callback:
        callbacks.append(callback)
    
    try:
        result = gp_minimize(
            objective_function_optimized, 
            refined_space, 
            n_calls=n_calls,
            n_initial_points=n_initial_points,
            x0=x0,  # Use starting point if provided
            callback=callbacks,
            random_state=42
        )
        
        # Save the final result
        from skopt.utils import dump
        dump(result, os.path.join(RUN_DIR, "final_result.pkl"))
        
        # Get the best result
        best_weights_array = result.x
        best_weights = {
            "incline_weight": best_weights_array[0],
            "max_incline_weight": best_weights_array[1],
            "distance_weight": best_weights_array[2],
            "zero_start_weight": best_weights_array[3]
        }
        
        # Save the best weights to a JSON file
        with open(os.path.join(RUN_DIR, "best_weights.json"), "w") as f:
            json.dump(best_weights, f, indent=2)
        
        if verbose >= 1:
            total_time = time.time() - TOTAL_START_TIME
            print("\nOptimization complete!")
            print(f"Total time: {format_time(total_time)}")
            print(f"Best weights found: {best_weights}")
            print(f"Best objective value: {result.fun}")
            print(f"All results and logs saved in: {RUN_DIR}")
            
            if start_weights:
                # Calculate improvement over starting weights
                start_value = result.func_vals[0] if x0 and len(result.func_vals) > 0 else None
                if start_value is not None:
                    improvement = (start_value - result.fun) / start_value * 100
                    print(f"Improvement over starting weights: {improvement:.2f}%")
        
        # Log the final result
        with open(LOG_FILE, "a") as f:
            f.write(f"\nOptimization completed at {datetime.datetime.now()}\n")
            f.write(f"Total time: {format_time(time.time() - TOTAL_START_TIME)}\n")
            f.write(f"Best weights: {best_weights}\n")
            f.write(f"Best objective value: {result.fun}\n")
            
            if start_weights:
                # Log improvement over starting weights
                start_value = result.func_vals[0] if x0 and len(result.func_vals) > 0 else None
                if start_value is not None:
                    improvement = (start_value - result.fun) / start_value * 100
                    f.write(f"Improvement over starting weights: {improvement:.2f}%\n")
        
        return result, best_weights, RUN_DIR
    
    except KeyboardInterrupt:
        print("\nOptimization interrupted by user!")
        print("Saving current state...")
        save_cache()
        print(f"You can resume later from checkpoint: {checkpoint_path}")
        return None, None, RUN_DIR
    
    except Exception as e:
        print(f"\nOptimization failed with error: {str(e)}")
        print("Saving current state...")
        save_cache()
        
        # Log the error
        with open(LOG_FILE, "a") as f:
            f.write(f"\nOptimization failed at {datetime.datetime.now()}\n")
            f.write(f"Error: {str(e)}\n")
        
        raise

# Simple sequential version that doesn't use multiprocessing 
# (useful for testing if all else fails)
def run_sequential(weights_dict=None, num_routes=10):
    """Run route simulations sequentially (no parallelism)"""
    if weights_dict is None:
        weights_dict = {
            "incline_weight": 4.0,
            "max_incline_weight": 4.0,
            "distance_weight": 10.0,
            "zero_start_weight": 10.0
        }
    
    start_time = time.time()
    
    # Process routes one at a time
    objective_results = []
    distance_results = []
    
    for i in range(min(num_routes, len(test_set))):
        result = simulate_single_route_optimized(i, weights_dict, 'both')
        if result['success']:
            objective_results.append(result['objective_consumption'])
            distance_results.append(result['distance_consumption'])
        print(f"Processed route {i+1}/{num_routes}")
    
    total_time = time.time() - start_time
    
    print(f"\nResults summary:")
    print(f"- Total time: {total_time:.2f} seconds")
    print(f"- Average time per route: {total_time/num_routes:.2f} seconds")
    print(f"- Average objective consumption: {sum(objective_results)/len(objective_results):.2f}")
    print(f"- Average distance consumption: {sum(distance_results)/len(distance_results):.2f}")
    
    return objective_results, distance_results


good_weights = {"incline_weight": 4.0, "max_incline_weight": 4.0, 
            "distance_weight": 10.0, "zero_start_weight": 10.0}

# Main function to run optimization with pre-optimized weights
def main():
    # Full optimization with all routes
    print("\n===== STARTING FULL OPTIMIZATION WITH PRE-OPTIMIZED WEIGHTS =====")
    result, best_weights, run_dir = optimize_bayesian_parallel(
        n_jobs=mp.cpu_count() - 1,  # Use all available cores except one
        n_calls=30,                 # Total number of evaluations
        n_initial_points=10,        # Initial exploration points
        num_routes=None,            # Use all routes in test_set
        verbose=2,                  # Detailed output
        log_dir="optimization_logs", # Directory to save results
        start_weights=good_weights  # Use the pre-optimized weights
    )
    
    # if result and best_weights:
    #     print("\nFinal evaluation of best weights with both routing types:")
    #     final_results = run_parallel_optimized(
    #         weights_dict=best_weights,
    #         num_routes=None,        # All routes
    #         calculation_type='both', # Evaluate both distance and objective routing
    #         n_jobs=mp.cpu_count() - 1,
    #         verbose=2
    #     )

# Uncomment the line below to run the optimization
main()
```
