```python
# Filter function to remove values outside +/- 7.5% of average
def filter_outliers(data_list, threshold_percent=15):
    avg = np.mean(data_list)
    lower_bound = avg * (1 - threshold_percent/100)
    upper_bound = avg * (1 + threshold_percent/100)
    
    print(f"Average: {avg:.2f}")
    print(f"Lower bound ({threshold_percent}% below avg): {lower_bound:.2f}")
    print(f"Upper bound ({threshold_percent}% above avg): {upper_bound:.2f}")
    
    
    filtered_list = [x for x in data_list if lower_bound <= x <= upper_bound]
    
    print(f"Original list length: {len(data_list)}")
    print(f"Filtered list length: {len(filtered_list)}")
    
    return filtered_list

# Import matplotlib and set the font to Times New Roman
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
from matplotlib import font_manager

# Set the font to Times New Roman for all text elements
plt.rcParams['font.family'] = 'Times New Roman'

# Set larger font sizes for all text elements
plt.rcParams['font.size'] = 14          # Base font size
plt.rcParams['axes.titlesize'] = 20     # Title font size
plt.rcParams['axes.labelsize'] = 16     # Axis label font size
plt.rcParams['xtick.labelsize'] = 14    # X-axis tick label font size
plt.rcParams['ytick.labelsize'] = 14    # Y-axis tick label font size
plt.rcParams['legend.fontsize'] = 14    # Legend font size

# Filter your lists
filtered_EOL_expected_dis_list = filter_outliers(EOL_expected_dis_list)
# filtered_EOL_expected_obj_list = filter_outliers(EOL_expected_obj_list)
adjusted_EOL_expected_dis_list = [value*0.5 for value in filtered_EOL_expected_dis_list]
print(f'Std.Dev: {np.std(filtered_EOL_expected_dis_list)}')
print(f'Co.Var: {(np.std(filtered_EOL_expected_dis_list)/np.average(filtered_EOL_expected_dis_list))*100:.2f}%')

# Create a larger figure with higher DPI for better quality when scaled
plt.figure(figsize=(16, 6), dpi=300)

# Increase marker size for better visibility
marker_size = 30

# plt.scatter(range(100), filtered_EOL_expected_dis_list[:100], s=marker_size,
#          label='Simulation Result')
plt.scatter(range(100), adjusted_EOL_expected_dis_list[:100], s=marker_size,color='red',
         label='Simulation Result')
# plt.scatter(range(100), filtered_EOL_expected_obj_list[:100], 
#          label='Simulation Result 2')

# Use thicker lines for better visibility
line_width = 3
# plt.axhline(y=np.average(filtered_EOL_expected_dis_list[:100]), color='orange', 
#            linestyle='--', linewidth=line_width, label='Average')
plt.axhline(y=np.average(filtered_EOL_expected_dis_list[:100])*0.5, color='green', 
           linestyle='--', linewidth=line_width, label='Average')

plt.title('Number of Cycles to 60% Capacity', fontweight='bold')
plt.xlabel('Simulation Number', fontweight='bold')
plt.ylabel('Number of Cycles', fontweight='bold')
plt.ylim(200, 800)

# Create a more visible grid
plt.grid(True, linestyle='-', linewidth=0.5, alpha=0.7)

# Make the legend more prominent with a border
plt.legend()

# Add a tight layout to use the figure area efficiently
# plt.tight_layout()

# Save with higher resolution
plt.savefig('validate_ageing.jpg', dpi=300, bbox_inches='tight')
plt.show()

```
