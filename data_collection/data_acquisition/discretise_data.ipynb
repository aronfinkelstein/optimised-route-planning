```python
with open("discretised_data.json", "r") as file:
    road_data = json.load(file)

def analyse_dis_data(data:dict):
    needs_el = 0
    total = 0
    i = 0

    for path, sections in data.items():  # Iterate over paths
        for section_name, section in sections.items():
                if "section" in section_name:
                    if section['climb'] == False:
                        coords = section['coords']
                        x1, y1 = coords[0], coords [1]
                        x2, y2 = coords[2], coords [3]
                        climb = ep.find_climb([x1,y1],[x2,y2])
                        inc_angle = ep.find_incline_angle(section['distance'], climb)
                        section['climb'] = climb
                        section['avg_incline_angle'] = inc_angle

                        print(f'section {i} complete')

                        i +=1
                        needs_el +=1
                        total += 1
                    else:
                        total += 1
                        i +=1
            
```
