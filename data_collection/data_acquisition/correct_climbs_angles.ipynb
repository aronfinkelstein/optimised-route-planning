```python

# Load the data
print("Loading data from '../data/large_stop_start_data.json'...")
with open('../data/large_stop_start_data.json', 'r') as f:
    map_data = json.load(f)

# Initialize counters for debugging
total_sections = 0
missing_data_sections = 0
processed_sections = 0

# Process each section
for outer_key, outer_value in map_data.items():
    print(f"\nProcessing group: {outer_key}")
    group_sections = 0
    group_fixed = 0
    
    for section_key, section_value in outer_value.items():
        if "section" in section_key:
            total_sections += 1
            group_sections += 1
            
            # Check if climb or avg_incline_angle is missing
            if not section_value['climb'] or not section_value['avg_incline_angle']:
                missing_data_sections += 1
                group_fixed += 1
                
                # Extract coordinates and distance
                coords = section_value['coords']
                distance = section_value['distance']
                
                print(f"  Fixing {section_key}:")
                print(f"    - Original data: climb={section_value['climb']}, angle={section_value['avg_incline_angle']}")
                print(f"    - Coords: {coords}")
                print(f"    - Distance: {distance}")
                
                try:
                    # Define points
                    point1 = [coords[0], coords[1]]
                    point2 = [coords[2], coords[3]]
                    
                    # Calculate missing values
                    climb = ep.find_climb(point1, point2)
                    angle = ep.find_incline_angle(distance, climb)
                    
                    # Update the data
                    section_value['climb'] = climb
                    section_value['avg_incline_angle'] = angle
                    
                    print(f"    - New values: climb={climb}, angle={angle}")
                    processed_sections += 1
                except Exception as e:
                    print(f"    - ERROR: {e}")
    
    print(f"  Group summary: {group_sections} sections, {group_fixed} sections fixed")

# Save the updated data to a new file
output_path = '../data/fixed_large_dis_data.json'
print(f"\nSaving updated data to '{output_path}'...")
with open(output_path, 'w') as f:
    json.dump(map_data, f, indent=2)

# Final summary
print("\nProcessing complete!")
print(f"Total sections: {total_sections}")
print(f"Sections with missing data: {missing_data_sections}")
print(f"Sections successfully processed: {processed_sections}")
print(f"Success rate: {(processed_sections / missing_data_sections * 100) if missing_data_sections > 0 else 0:.2f}%")
```
